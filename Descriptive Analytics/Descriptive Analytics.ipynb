```python
import pandas as pd
import numpy as np

data = [10, 15, 14, 10, 10, 18, 25, 30, 40]
df = pd.DataFrame(data, columns=['Values'])
print(df.head())

mean_value = df['Values'].mean()
median_value = df['Values'].median()
mode_value = df['Values'].mode()[0]

print(f"Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")
```

```python
std_dev = df['Values'].std()
variance = df['Values'].var()
iqr = df['Values'].quantile(0.75) - df['Values'].quantile(0.25)

print(f"Standard Deviation: {std_dev}, Variance: {variance}, IQR: {iqr}")
```

```python
from scipy.stats import skew, kurtosis, shapiro

skewness = skew(df['Values'])
kurt = kurtosis(df['Values'])
shapiro_test = shapiro(df['Values'])

print(f"Skewness: {skewness}, Kurtosis: {kurt}, Shapiro-Wilk: {shapiro_test}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Values'], bins = 10, kde = True)
plt.title("Histogram of Values")
plt.show()

sns.boxplot(y = df['Values'])
plt.title("Boxplot for Outlier Detection")
plt.show()
```
